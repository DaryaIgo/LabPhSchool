## 📚 Базовая теория и формулы

При **равномерном** движении по окружности модуль скорости постоянен, но направление меняется. Поэтому существует **центростремительное ускорение** $\vec{a}$, направленное к центру.

### Основные величины

| Величина | Обозначение | Формула | Единица СИ |
|----------|-------------|---------|------------|
| Период обращения | $T$ | $T = \frac{t}{N}$ | с |
| Частота обращения | $\nu$ | $\nu = \frac{1}{T} = \frac{N}{t}$ | Гц |
| Угловая скорость | $\omega$ | $\omega = \frac{2\pi}{T} = 2\pi\nu$ | рад/с |
| Линейная (орбитальная) скорость | $v$ | $v = \omega R = \frac{2\pi R}{T}$ | м/с |
| Центростремительное ускорение | $a$ | $a = \frac{v^2}{R} = \omega^2 R$ | м/с² |
| Угол поворота (при равномерном) | $\varphi$ | $\varphi = \omega t$ | рад |

### Связь линейной и угловой скорости
$$v = \omega R$$

### Важно помнить
- Угловая скорость $\omega$ одинакова для всех точек твёрдого тела, вращающегося вокруг неподвижной оси.
- Линейная скорость $v$ тем больше, чем дальше точка от оси вращения ($v \propto R$).


## ⚠️ Лайфхаки и «ловушки» экзамена

> **1. Период vs частота**  
> Частота $\nu$ — это количество оборотов в секунду. Если в задаче дано число оборотов **в минуту** ($n$ об/мин), обязательно переведите в Гц:  
> $$\nu = \frac{n}{60}$$

> **2. Радиус vs диаметр**  
> В условии часто дают диаметр $D$ окружности. Не забудьте: $R = \frac{D}{2}$. Подстановка $D$ вместо $R$ — классическая ошибка.

> **3. Центростремительное ускорение есть всегда**  
> Даже если модуль скорости постоянен, при движении по кривой **обязательно** есть ускорение, перпендикулярное скорости. При равномерном движении по окружности тангенциальное ускорение равно нулю ($a_{\tau} = 0$), и остаётся только центростремительное.

> **4. Сцепленные колёса и ремни**  
> - Если два колеса соединены **ремнём без скольжения**, их линейные скорости на ободе равны: $v_1 = v_2$.  
> - Если колёса **сцеплены** (зубчатые шестерни), то $v_1 = v_2$, но направления вращения противоположны.  
> - Для точек на **одном твёрдом теле** угловые скорости равны: $\omega_1 = \omega_2$.

> **5. Единицы угловой скорости**  
> В задачах иногда встречается $\omega$ в градусах в секунду. Переводите в радианы: $1^{\circ} = \frac{\pi}{180}$ рад, либо сразу используйте $\omega = \frac{\Delta\varphi}{t}$, где $\Delta\varphi$ в радианах.


## 📝 Задача 1. Карусель (базовый уровень)

**Условие:**  
Ребёнок сидит на карусели на расстоянии $R = 3{,}5$ м от оси вращения. Карусель совершает 5 оборотов за $t = 25$ с.

**Найдите:**
1. Период и частоту вращения.
2. Линейную скорость ребёнка.
3. Центростремительное ускорение.

**Аналитический разбор:**  
Задача прямого применения формул. Сначала находим период $T = t/N$, затем частоту $\nu = 1/T$. Далее по радиусу и периоду вычисляем $v = 2\pi R / T$, а ускорение — $a = v^2 / R$.


````python
import math

# --- Исходные данные ---
R = 3.5        # радиус окружности, м
N = 5          # число полных оборотов
t = 25         # время, с

# --- 1. Период и частота ---
# Период — время одного оборота
T = t / N
# Частота — число оборотов в секунду
nu = 1 / T

print(f"1) Период вращения: T = {T:.1f} с")
print(f"   Частота вращения: ν = {nu:.2f} Гц")

# --- 2. Линейная скорость ---
# v = 2πR / T
v = 2 * math.pi * R / T
print(f"2) Линейная скорость: v = {v:.2f} м/с")

# --- 3. Центростремительное ускорение ---
# a = v² / R
a = v**2 / R
print(f"3) Центростремельное ускорение: a = {a:.2f} м/с²")
````

## 📝 Задача 2. Автомобиль на повороте (средний уровень)

**Условие:**  
Автомобиль движется по изогнутому участку дороги радиусом $R = 50$ м со скоростью $v = 54$ км/ч.

**Найдите:**
1. Угловую скорость автомобиля.
2. Центростремительное ускорение.

**Аналитический разбор:**  
Сначала переведём скорость из км/ч в м/с: $1$ км/ч $= \frac{1000}{3600}$ м/с $= \frac{1}{3{,}6}$ м/с.  
Угловая скорость находится из связи $v = \omega R \;\Rightarrow\; \omega = v/R$.  
Ускорение — по формуле $a = v^2 / R$ (или $a = \omega^2 R$ для проверки).


````python
import math

# --- Исходные данные ---
R = 50            # радиус поворота, м
v_kmh = 54        # скорость, км/ч

# --- Перевод скорости в СИ (м/с) ---
# 1 км/ч = 1000 м / 3600 с = 1/3.6 м/с
v = v_kmh / 3.6
print(f"Скорость в СИ: v = {v:.1f} м/с")

# --- 1. Угловая скорость ---
# ω = v / R
omega = v / R
print(f"1) Угловая скорость: ω = {omega:.3f} рад/с")

# --- 2. Центростремительное ускорение ---
# a = v² / R
a = v**2 / R
print(f"2) Центростремельное ускорение: a = {a:.2f} м/с²")

# --- Проверка: a = ω² * R ---
a_check = omega**2 * R
print(f"   Проверка (ω²R): a = {a_check:.2f} м/с²")
````

## 📝 Задача 3. Зубчатые шестерни (повышенный уровень, ЕГЭ)

**Условие:**  
Две зубчатые шестерни сцеплены друг с другом. Радиус первой шестерни $R_1 = 10$ см, второй — $R_2 = 25$ см.  
Угловая скорость первой шестерни $\omega_1 = 5$ рад/с.

**Найдите:**
1. Угловую скорость второй шестерни $\omega_2$.
2. Отношение центростремительных ускорений точек на ободах $\frac{a_1}{a_2}$.

**Аналитический разбор:**  
При сцеплении зубчатых колёс линейные скорости точек на ободах равны: $v_1 = v_2$.  
Отсюда $\omega_1 R_1 = \omega_2 R_2$, следовательно $\omega_2 = \omega_1 \frac{R_1}{R_2}$.  
Центростремительное ускорение $a = \omega^2 R = \omega v$. Поскольку $v_1 = v_2$, удобно использовать $a = \frac{v^2}{R}$, откуда $\frac{a_1}{a_2} = \frac{R_2}{R_1}$.


````python
import math

# --- Исходные данные ---
R1_cm = 10        # радиус первой шестерни, см
R2_cm = 25        # радиус второй шестерни, см
omega1 = 5        # угловая скорость первой шестерни, рад/с

# --- Перевод радиусов в СИ (м) ---
R1 = R1_cm / 100
R2 = R2_cm / 100

# --- Линейная скорость на ободе первой шестерни ---
v1 = omega1 * R1

# При сцеплении линейные скорости равны: v1 = v2
v2 = v1

# --- 1. Угловая скорость второй шестерни ---
# ω1*R1 = ω2*R2  =>  ω2 = ω1 * R1 / R2
omega2 = omega1 * R1 / R2
print(f"1) Угловая скорость второй шестерни: ω₂ = {omega2:.2f} рад/с")

# --- 2. Отношение центростремительных ускорений ---
# a = v² / R, при v1 = v2  =>  a1/a2 = (v²/R1) / (v²/R2) = R2 / R1
a_ratio = R2 / R1
print(f"2) Отношение ускорений: a₁ / a₂ = {a_ratio:.2f}")

# Проверка через прямой расчёт
a1 = v1**2 / R1
a2 = v2**2 / R2
print(f"   Проверка: a₁ = {a1:.3f} м/с², a₂ = {a2:.3f} м/с²")
print(f"   a₁ / a₂ = {a1/a2:.2f}")
````

## 🎨 Визуализация: движение по окружности

Ниже приведён код, строящий окружность, положение точки, а также векторы скорости и центростремительного ускорения.


````python
import numpy as np
import matplotlib.pyplot as plt

# Параметры окружности
R = 2.0           # радиус, м (условные единицы для графика)
omega = 1.0       # рад/с
t = 1.2           # момент времени, с

# Положение точки
phi = omega * t   # угол поворота
x = R * np.cos(phi)
y = R * np.sin(phi)

# Вектор скорости (касательный, перпендикуляр радиусу)
v = omega * R
vx = -v * np.sin(phi)
vy = v * np.cos(phi)

# Вектор центростремительного ускорения (к центру)
a = omega**2 * R
ax = -a * np.cos(phi)
ay = -a * np.sin(phi)

# Построение графика
fig, ax_plot = plt.subplots(figsize=(6, 6))
theta = np.linspace(0, 2*np.pi, 200)
ax_plot.plot(R*np.cos(theta), R*np.sin(theta), 'b--', label='Траектория')

# Точка
ax_plot.plot(x, y, 'ro', markersize=8, label='Материальная точка')

# Вектор скорости (зелёный)
ax_plot.quiver(x, y, vx, vy, color='green', scale=5, width=0.006,
               label='Скорость $\\vec{v}$')

# Вектор ускорения (красный)
ax_plot.quiver(x, y, ax, ay, color='red', scale=15, width=0.006,
               label='Ускорение $\\vec{a}$')

# Оформление
ax_plot.set_aspect('equal')
ax_plot.axhline(0, color='black', linewidth=0.5)
ax_plot.axvline(0, color='black', linewidth=0.5)
ax_plot.set_xlim(-3, 3)
ax_plot.set_ylim(-3, 3)
ax_plot.set_title('Равномерное движение по окружности')
ax_plot.legend(loc='upper right')
plt.grid(True)
plt.show()
````
